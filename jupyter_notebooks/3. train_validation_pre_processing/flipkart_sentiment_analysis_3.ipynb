## 4.1 New Sentiments column
  > The below used method is called **method chaining**.<br>
  > We will use it wherever possible.


### 4.1.1 We can create new column "sentiments" using numpy in 3 ways,

#### 4.1.1.1 First way

- Create
  - Variable.
  - Conditions.
  - Choices.

```python
col         = "ratings"
conditions  = [ flipkart_elec_mob[col] > 3, flipkart_elec_mob[col] == 3, flipkart_elec_mob[col] < 3 ]
choices     = [ "positive", "neutral", "negative" ]
```

- Create new sentiment column.

```python
(
    flipkart_elec_mob
    .drop(columns=["sentiments"])
    .assign(
              sentiments = np.select(conditions, choices, default=np.nan)
           )
    .head()
)
```

#### 4.1.1.2 Second way
- Create new sentiment column.

```python
(
    flipkart_elec_mob
    .drop(columns=["sentiments"])
    .assign(
              sentiments = np.where(
                                      flipkart_elec_mob["ratings"] > 3, "positive",
                                      np.where(
                                                flipkart_elec_mob["ratings"] < 3, "negative", "neutral"
                                              )
                                   )
           )
    .head()
)
```

***Observations:***
- If you have complex multiple condition, then this can be a big issue.

#### 4.1.1.3 Third way
- Create a function.

```python
def conditions(x):
    if   x > 3:   return "positive"
    elif x < 3:   return "negative"
    else:         return "neutral"
```

- Use numpy.vectorize(),
  - Vectorized operations in NumPy enable the use of   efficient, pre-compiled functions and mathematical<br> operations on NumPy arrays and data sequences.
  - Vectorization is a method of performing array operations without the use of for loops.

```python
func = np.vectorize(conditions)
```

- Create new sentiment column.

```python
(
    flipkart_elec_mob
    .drop(columns=["sentiments"])
    .assign(
              sentiments = func(flipkart_elec_mob["ratings"])
           )
    .head()
)
```

### 4.3.5 `lemmatize_func`

- `CC`: Coordinating conjunction (e.g., and, but)
- `CD`: Cardinal digit (e.g., one, two)
- `DT`: Determiner (e.g., the, a)
- `EX`: Existential 'there' (e.g., 'There is a problem')
- `FW`: Foreign word (e.g., caf√©, t√™te-√†-t√™te)
- `IN`: Preposition or subordinating conjunction (e.g., in, on, because)
- `JJ`: Adjective (e.g., big, small)
- `JJR`: Comparative adjective (e.g., bigger, smaller)
- `JJS`: Superlative adjective (e.g., biggest, smallest)
- `LS`: List marker (e.g., 1), 2))
- `MD`: Modal verb (e.g., could, will)
- `NN`: Noun, singular (e.g., desk, car)
- `NNS`: Noun, plural (e.g., desks, cars)
- `NNP`: Proper noun, singular (e.g., Harrison, Paris)
- `NNPS`: Proper noun, plural (e.g., Americans, Beatles)
- `PDT`: Predeterminer (e.g., all the kids, both of them)
- `POS`: Possessive ending (e.g., parent's, dog's)
- `PRP`: Personal pronoun (e.g., I, he, she)
- `PRP$`: Possessive pronoun (e.g., my, his, hers)
- `RB`: Adverb (e.g., very, silently)
- `RBR`: Comparative adverb (e.g., better, worse)
- `RBS`: Superlative adverb (e.g., best, worst)
- `RP`: Particle (e.g., give up, break down)
- `TO`: The word 'to' (e.g., to go, to see)
- `UH`: Interjection (e.g., errrrrrrrm, wow)
- `VB`: Verb, base form (e.g., take, run)
- `VBD`: Verb, past tense (e.g., took, ran)
- `VBG`: Verb, gerund or present participle (e.g., taking, running)
- `VBN`: Verb, past participle (e.g., taken, run)
- `VBP`: Verb, singular present, non-3rd person (e.g., take, run)
- `VBZ`: Verb, 3rd person singular present (e.g., takes, runs)
- `WDT`: Wh-determiner (e.g., which, that)
- `WP`: Wh-pronoun (e.g., who, what)
- `WP$`: Possessive wh-pronoun (e.g., whose)
- `WRB`: Wh-adverb (e.g., where, when)


```python
def lemmatize_func(x):
  words = []
  for i in range(len(x)):
    word = x[i][0]
    tag = x[i][1]
    if tag in ["VB", "VBD", "VBG", "VBN", "VBP", "VBZ"]:
      pos = "v"
      words.append(WordNetLemmatizer().lemmatize(word, pos=pos))
    elif tag in ["JJ", "JJR", "JJS"]:
      pos = "a"
      words.append(WordNetLemmatizer().lemmatize(word, pos=pos))
    elif tag in ["RB", "RBR", "RBS"]:
      pos = "r"
      words.append(WordNetLemmatizer().lemmatize(word, pos=pos))
    else:
      pos = "n"
      words.append(WordNetLemmatizer().lemmatize(word, pos=pos))

  return " ".join(words)
```

### 4.4.4 Extract `emoticons` and `emotions`

#### 4.4.4.1 Create extract_emotions function

```python
def extract_emotions():
  emoticons = []
  emotions = []
  for i in range(1, len(soup.find_all('tr'))):
    emote = soup.find_all('tr')[i].td.text
    emoticons.append(emote)
    emotion = soup.find_all('tr')[i].find_all("td")[-2].text.lower()
    emotion = emotion.replace(" ", "-")
    emotion = " "+emotion+" "
    emotions.append(emotion)
  return emoticons, emotions
```

### 5.3.3 Extract `emoticons` and `emotions`

#### 5.3.3.1 Create extract_emotions function

```python
def extract_emotions():
  emoticons = []
  emotions = []
  for i in range(1, len(soup.find_all('tr'))):
    emote = soup.find_all('tr')[i].td.text
    emoticons.append(emote)
    emotion = soup.find_all('tr')[i].find_all("td")[-2].text.lower()
    emotion = emotion.replace(" ", "-")
    emotion = " "+emotion+" "
    emotions.append(emotion)
  return emoticons, emotions
```

#### 5.3.3.2 `extract_emotion`

```python
%%time
emoticons, emotions = extract_emotions()
```

## 5.4 Stopwords removed - Yes

### 5.4.1 `stopwords_remove`

```python
def stopwords_remove(x):
  stop_words = set(stopwords.words('english'))
  words = [word for word in x.split() if word not in stop_words]
  words = " ".join(words)
  return words
```

```python

flipkart_electronics = (
                      flipkart_electronics
                      .assign
                      (
                          no_stopwords_reviews = flipkart_electronics["yes_emoji_reviews"]
                          .apply(stopwords_remove)
                      )
                     )
```
